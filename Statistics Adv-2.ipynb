- &sigma; = standard deviation
- &mu; = mean

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans. Cumulative Density Function (CDF) of a random variable X at a given point x, is a function that gives the probability that it takes a value less than or equal to x.
$$CDF(x)=P(X<=x)$$
The CDF always starts at 0 and ends at 1.

Example- Suppose we throw a dice, then probability of occurence of each number is 1/6, i.e, P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=1/6.
Now CDF is calculated as sum of probabilities of points lower than or equal to a given point. For example, CDF(3)= P(X<=3) = P(1)+P(2)+P(3) = 1/6+1/6+1/6 = 3/6 =0.5. Similarly for all values:$P(X\le{x})$

Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

Ans. Normal distribution can often be used in situations where we have a lot of data. Examples are, Height and Weight, IQ Scores, Measurement Errors, Exam Scores, Financial Returns, Natural Phenomena, Quality Control, Biological Traits, Test Scores, Consumer Behavior, Agricultural Yields, Disease Incidence, Customer Satisfaction Ratings, etc.

The shape of normal distribution depends upon its parameters mean (&mu;) and standard deviation (&sigma;) :

- Mean (&mu;) - The peak of normal distribution occurs at mean. The mean is point of highest probability and the distribuiton is symmetric around mean. If mean is shiftedd to left or right, then the entire distribution shifts to left or right. 

- Standard deviation (&sigma;) - It gives spread of distribution. Larger value of &sigma; gives a flat and wide distribution and smaller value gives a narrow and tall distribution.

Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

Ans. **Bernoulli Distribution** is used in cases where we have a binary outcomes. Here we have a discrete random variable. Every experiment has only 2 possible outcomes k={0,1}. The probability of success is p and of failure is q=1-p. Example, tossing of coin one time, student passing or failing.

>PMF $$ P (k,p) =  p^k.(1-p)^{1-k} $$ where k = {0,1}
>$$ P(k=1) = p $$
>$$ P(k=0) = (1-p) = q $$
>Mean $$ \mu = p $$

**Binomial distribution** is a series of individual Bernoulli trials with binary outcomes. Each trial has only 2 possible outcomes, Success or Failure. It has 2 parameters, n and p. n is the total number of trials and p is probability of success. Probability of success is q=1-p. Examples where Binomial Distribution can be applied are tossing a coin again and again as a coin has only 2 outcomes and winning or loosing of a team or player in a series of matches.

>PMF $$ P (k,n,p) =  {}^{n}C_{k}.p^k.(1-p)^{n-k} $$
>Mean $$ \mu = np $$


Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.

Ans. Given, &mu; = 50, &sigma; = 10, we have to find probability of X>60.
The Z-score can be calculated as:
$$ Z = \frac{X-\mu}{\sigma} $$

$$ Z = \frac{60 - 50}{10} $$

$$ Z = \frac{10}{10} $$

$$ Z = 1.00 $$

Now for this Z score, the P-value value from Z-stat table is 0.8413, this gives area of curve below X=60, thus area of curve above X=60 is 1 - 0.8413 = 0.1587. 

Thus probability that a randomly selected observation will be greater than 60 is 0.1587 or 15.87%.

Q8: What is the z score? State the importance of the z score.

Ans. The Z score is a statiscal meassure that tells us how many standard deviations a data point is away from mean. It is calculated as :
$$ Z = \frac{X-\mu}{\sigma} $$
where, 
- X = point at which Z score is calculated
- &mu; = mean
- &sigma; = standard deviation

A positive Z score (Z>0) means data point is at right side of mean, negative (Z<0) means it is on left side of mean and at mean, Z=0.

The Z score is very important, it helps us in Hypothesis testing, in checking wheter our Null hypothesis is trtue or false. Using a Z stat table, we can also calculate the p value of a Z score. P value gives the area under curve on left side of a point, p value gives probability.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans. **Central Limit Theorem-** The sampling distribution of the sample mean (or the sum) of a sufficiently large number of independent, identically distributed random variables will be normally distributed, regardless of whether the population has a Normal, Poisson, Bernoulli, or any other distribution. 

**Significance-** The Central Limit Theorem is particularly useful in situations where we're dealing with sample means, sums, or averages of data. It provides a bridge between the characteristics of a sample and the characteristics of the underlying population, even when the original population may not follow a normal distribution. CLT is used in various types of Hypothesis testing techniques, confidence interval measurements,etc

Q10: State the assumptions of the Central Limit Theorem.

Ans. Basic assumptions of Central Limit Theorem are:

- **Independence** - The random variables should be independent of each other. Value of one variable should not affect the value of another.
- **Identically Distributed** - The random variables should be drawn from the same underlying population distribution. They should have the same mean (&mu;) and the same standard deviation (&sigma;).
- **Sample Size** - The sample size is large enough. Larger sample size leads to better results. In general n$\ge$30.
- **Randomness** - Samples are collected randomly, there should be no bias.
- **Outlier** - Outliers should not be there, their presence can affect the statistical result
