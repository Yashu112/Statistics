Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

Ans. We know that in normal distribution, formula of probability distribution function is
> $$ PDF = \frac{1}{\sigma\sqrt{2\pi}} e^{{\frac{-1}{2}} \left(\frac{x_i - \mu}{\sigma}\right)^2} $$ 

The code will be as ahown:

```python
def pdf(x, mean, std):
    from numpy import sqrt, pi, exp
    
    denom = std * sqrt(2*pi)
    sq = ((x - mean)/std)**2
    pdf =  exp(-0.5 * sq)/denom
    return pdf
```

Example of application

```python
import numpy as np
import seaborn as sns

data = np.random.normal(size=50)

sns.kdeplot(data)
sns.set_style('whitegrid')
```

```python
mean = np.mean(data)
std = np.std(data)
```

From above graph, density is around 0.20 at point 1, let us use our function to calculate it and check if it works well

```python
pdf(1,mean,std)
```

Thus we see that the function created works well.

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

Ans. Properites of binomial distribution:
- it is a pmf, Probability Mass Function
- it is a discrete probability distribution
- it is a series of individual Bernoulli trial
- each trial has only 2 possible outcomes, Success or Failure
- it has 2 parameters, n and p
    - n is the total number of trials
    - p is probability of success
- notation : `B(n,p)`

suppose k = {0,1,2,3, ... } is the number of successes, then

- PMF: $$ P (k,n,p) =  {}^{n}C_{k}.p^k.(1-p)^{n-k} $$
    where, $$  {}^{n}C_{k} = \frac{n!}{k!(n-k)!} $$
- Mean: $np$
- Varaiance: $var = npq$
- Stdandard Deviation: $std = \sqrt{npq}$

Two Examples where Binomial Distribution can be applied:
- Tossing a coin again and again as a coin has only 2 outcomes
- Winning or loosing of a team or player in a series of matches

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

Ans. We know that for poisson distribution:

$$ {\displaystyle cdf = e^{-\lambda }\sum _{j=0}^{\lfloor k\rfloor }{\frac {\lambda ^{j}}{j!}}} $$

Let us define a function as shown:

```python
def cdf(x, mean):
    from math import factorial as f
    from math import floor, exp
    
    sum = 0 
    for i in range(0,floor(x)):
        sum = sum + (mean**i)/f(i)
        
    return exp(-mean)*sum
```

```python
from math import factorial as f
f(5)
```

```python
import numpy as np
pois = np.random.poisson(5.0, size=1000)
np.mean(pois)
```

Let's draw a cdf plot of above poisson distribution data:

```python
import seaborn as sns
sns.kdeplot(pois, cumulative=True)
sns.set_style('whitegrid')
```

From above plot, density at point 8 is between 0.8 and 1.0, let's use our formula

```python
cdf(7.5, 5)
```

```python
from scipy.stats import poisson
poisson.cdf(7.5,5)
```

So our function works well

Q7. How Binomial distribution different from Poisson distribution?

Ans. Both Binomial and Poisson distribution are pmf, Probability Mass Functions and are a discrete probability distribution.

Binomial distribution is a series of individual Bernoulli trials with binary outcomes. Each trial has only 2 possible outcomes, Success or Failure. It has 2 parameters, n and p. n is the total number of trials and p is probability of success. Probability of success is q=1-p. Examples where Binomial Distribution can be applied are tossing a coin again and again as a coin has only 2 outcomes and winning or loosing of a team or player in a series of matches.

>PMF $$ P (k,n,p) =  {}^{n}C_{k}.p^k.(1-p)^{n-k} $$

Poisson Distribution is used to represent number of events occuring in a fixed time interval. For example, number of peoplke visiting a bank each hour, or number of vehicles passing a particular location or number of visitors of restaurant every hour.

>PMF $$ P(\lambda,k) = \frac{e^{-\lambda}.\lambda^k}{k!} $$
